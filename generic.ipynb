```python
from concurrent.futures import ThreadPoolExecutor
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

def open_and_extract(link):
    driver = webdriver.Chrome()
    try:
        driver.get(link)
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, 'p')))
        
        page_content = driver.page_source
        soup = BeautifulSoup(page_content, 'html.parser')
        p_tags = soup.find_all('p')
        p_contents = [p.get_text() for p in p_tags]
        print(f"Page: {link}\n<p> tag contents: {p_contents}\n{'='*30}")
    finally:
        driver.quit()

if __name__ == "__main__":
    driver = webdriver.Chrome()
    driver.get("https://www.google.com")

    search_box = driver.find_element(By.XPATH, '//*[@id="APjFqb"]')
    search_box.send_keys("Insurance")
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="APjFqb"]')))

    href_elements = driver.find_elements(By.XPATH, '//a[@href]')
    href_list = [element.get_attribute('href') for element in href_elements]


    driver.quit()
    num_threads = 10

    with ThreadPoolExecutor(max_workers=num_threads) as executor:
        # Use executor.map to apply the open_and_extract function to each link in parallel
        executor.map(open_and_extract, href_list)

```
