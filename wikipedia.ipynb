```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

driver2 = webdriver.Chrome()
driver2.get("https://www.wikipedia.org/")

search_box = driver2.find_element(By.ID, "searchInput")
search_box.send_keys("Insurance")

# To wait for the dropdown to appear and then store it
search_results = WebDriverWait(driver2, 15).until(
    EC.presence_of_all_elements_located((By.CLASS_NAME, "suggestions-dropdown"))
)



for i in range(1, min(10, len(search_results) + 1)):
    result = search_results[i - 1]  # here search result stores multiples items in the kind of list structure
    

    #link_element = result.find_element(By.XPATH, f'//*[@id="typeahead-suggestions"]/div/a[{i}]')
    #print(link_element.text)
    #link_element.click()
    result.click()

    soup = BeautifulSoup(driver2.page_source, 'html.parser')
    paragraphs = soup.find_all("p")

    for paragraph in paragraphs:
        print(paragraph.get_text(strip=True))

    driver2.back()

driver2.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

driver3 = webdriver.Chrome()
driver3.get("https://www.wikipedia.org/")

search_box = driver3.find_element(By.ID, "searchInput")
search_box.send_keys("Insurance")

# Wait for the suggestions dropdown to appear
search_results = WebDriverWait(driver3, 10).until(
    EC.presence_of_all_elements_located((By.XPATH, '//*[@id="typeahead-suggestions"]/div/a'))
)

search_count = len(search_results)

for i in range(1, search_count + 1):
    search_result = WebDriverWait(driver3, 10).until(
        EC.presence_of_element_located((By.XPATH, f'//*[@id="typeahead-suggestions"]/div/a[{i}]'))
    )
    
    link_text = search_result.text
    
    print(f"Result {i} Text: {link_text}")
    
    search_result.click() # search_result -> here it is only storing one data tha's why we cannot iterarte over it like before

    soup = BeautifulSoup(driver3.page_source, 'html.parser')
    paragraphs = soup.find_all("p")

    for paragraph in paragraphs:
        print(paragraph.get_text(strip=True))

    print(f"Clicked on result {i}: {link_text}")
    driver3.back()

driver3.quit()
```

```python
import pandas as pd
import requests
from io import BytesIO

excel_url = 'https://irdai.gov.in/documents/37343/367479/November+2022+%28Line+of+Business+wise+data%29.xlsx/183d9a29-a7d8-572f-48a0-ad3afa01618d?version=1.0&t=1670476245273&download=true'
response = requests.get(excel_url)

if response.status_code == 200:
    excel_data = pd.ExcelFile(BytesIO(response.content)) # Reaing the content of the response into a Pandas DataFrame

    sheet_names = excel_data.sheet_names

    # dictionary to store DataFrames, where keys are names of the sheets
    dfs = {}
    for sheet_name in sheet_names:
        dfs[sheet_name] = excel_data.parse(sheet_name, header=None) 

    for sheet_name, df in dfs.items():
        print(f"Sheet: {sheet_name}")
        print(df.head())
        print("\n" + "="*30 + "\n")
else:
    print(f"Failed to fetch the file. Status code: {response.status_code}")
```
